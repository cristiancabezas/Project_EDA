```python
tabla = []

for x in filas_individuales:
    individual = x.find_elements(By.TAG_NAME, 'td')
    fila_entera = []
    
    for i in individual:
        valor = i.text
        fila_entera.append(valor)

    tabla.append(fila_entera)
```

```python
tabla
```

```python
columnas = ['Fecha', 'Renta Bruta', 'Borrar', 'Renta Disponible', 'Puesto Nacional', 'Puesto CCAA', 'Poblacion']

df_renta = pd.DataFrame(tabla, columns=columnas)
```

```python
df_renta.drop(columns = ['Puesto Nacional', 'Puesto CCAA'], inplace= True)
```

```python
import re
def eliminar_euro(col):
    return re.sub(r'\s\€', '', col)
```

```python
df_renta['Renta Bruta'] = df_renta['Renta Bruta'].apply(eliminar_euro)
```

```python
def quitar_comas(col):
    return col.replace('.','')
```

```python
df_renta['Renta Bruta'] = df_renta['Renta Bruta'].astype('object')
```

```python
df_renta[['Renta Bruta','Renta Disponible','Poblacion']] = df_renta[['Renta Bruta','Renta Disponible','Poblacion']].apply(quitar_comas)
```

```python
df_renta['Renta Bruta'] = df_renta['Renta Bruta'].replace('.','fdzg')
```

```python
df_renta.info()
```

```python
df_renta['Renta Bruta'] = df_renta['Renta Bruta'].apply(lambda x : str(x).replace('.',''))
df_renta['Renta Disponible'] = df_renta['Renta Disponible'].apply(lambda x : str(x).replace('.',''))
df_renta['Poblacion'] = df_renta['Poblacion'].apply(lambda x : str(x).replace('.',''))
```

```python
df_renta[['Renta Bruta','Renta Disponible','Poblacion']] = df_renta[['Renta Bruta','Renta Disponible','Poblacion']].astype(int)
```

```python
df_renta.info()
```

```python
df_renta['Renta_per_capita'] = df_renta['Renta Disponible'] / df_renta['Poblacion']
```

```python
df_renta['Renta_bruta_per_capita'] = df_renta['Renta Bruta'] / df_renta['Poblacion']
```

```python
info_locales = []

for comercio in selector_casilla_restaurante:
    local = []
    
    try:
        nombre = comercio.find_element(By.CLASS_NAME, 'NrDZNb').text
    except:
        nombre = ''
    local.append(nombre)

    try:
        estrellas = comercio.find_element(By.CLASS_NAME, 'MW4etd').text
    except:
        estrellas = ''
    local.append(estrellas)

    try:
        reseñas = comercio.find_element(By.CLASS_NAME, 'UY7F9').text
    except:
        reseñas = ''
    local.append(reseñas)

    try:
        precio = comercio.find_element(By.CLASS_NAME, 'AJB7ye').text
    except:
        precio = ''
    local.append(precio)

    # Reemplazamos los XPATH por búsqueda por clase y posición
    try:
        detalles = comercio.find_elements(By.CLASS_NAME, 'W4Efsd')
        tipo_local = detalles[1].text
        direccion = detalles[2].text 
    except:
        tipo_local = ''
        direccion = ''
    
    local.append(tipo_local)
    local.append(direccion)

    info_locales.append(local)
```

```python
def quitar_info_despues_de_punto(col):
    if col.count('·') >=1:
        expresion = re.search(r'^(.*?)(?= ·)', col)
        if expresion:
            return expresion.group(1)
        else:
            return col
    else:
        return col
```
