```python
df_tipificacion_renta['Clasificada_como'] = 'Categorica'
df_tipificacion_renta.loc[df_tipificacion_renta.Card == 2, 'Clasificada_como'] = 'Binaria'
df_tipificacion_renta.loc[df_tipificacion_renta.Card > 10, 'Clasificada_como'] = 'Numerica Discreta'
df_tipificacion_renta.loc[df_tipificacion_renta['%_Card'] > 30, 'Clasificada_como'] = 'Numerica Continua'
df_tipificacion_renta
```

|Nombre|Descripción|Tipo de variable|
|------|-----------|----------------|
|Fecha|Fecha por años|Numerica Continua|
|Renta bruta|El dinero antes de impuestos|Numerica Continua|
|Renta disponible|El dinero que te queda después de impuestos|Numerica Continua|
|Población|La cantidad de personas por año|Numerica Continua|
|Renta per capita|Es la renta que queda después de impuestos por persona|Numerica Continua|
|Renta bruta per capita|Es la renta antes de impuestos por persona|Numerica Continua|

```python
df_renta
```

```python
fig, ax = plt.subplots()
ax.plot(df_renta['Fecha'], df_renta['Renta Bruta'], label='Renta Bruta')
ax.plot(df_renta['Fecha'], df_renta['Renta Disponible'], label='Renta Disponible')
ax.set_ylabel('Renta Bruta y Disponible')
ax.legend()
ax.set_title('Renta Badalona')

ax_capita = ax.twinx() #creamos un segundo eje Y
ax_capita.plot(df_renta['Fecha'], df_renta['Renta_per_capita'], label='Renta per cápita', color = 'green' ,linestyle='--')
ax_capita.set_ylabel('Renta per cápita')
ax_capita.legend(loc = 'upper right');
```

```python
# como vemos la renta es ascendente, siempre hay una separación entre la bruta y la disponible, eso refleja los impuestos que pagamos,
# luego la renta per cápita también es ascendente, se ve una subida notable después del 2020, seguramente al haber mas teletrabajo
# las empresas podrían pagar mas y por eso aumentar la renta per cápita
```

```python
# Renta Bruta: El ingreso antes de impuestos
# Renta Disponible: Dinero despues de impuestos, lo que tienes a tu disposición despues de todo
# Renta per Cápita: El ingreso promedio por persona
```

```python
correlacion_bruta = df_renta['Poblacion'].corr(df_renta['Renta_bruta_per_capita'])
correlacion_disp = df_renta['Poblacion'].corr(df_renta['Renta_per_capita'])
print(correlacion_bruta)
print(correlacion_disp)
```

```python
#Como se ve con los numeros en la correlacion, a mayor población, mayor renta per cápita
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2, 1)
sns.scatterplot(df_renta, x='Poblacion', y = 'Renta_bruta_per_capita')
plt.title('Población VS renta bruta per cápita')

plt.subplot(1,2,2)
sns.scatterplot(df_renta, x='Poblacion', y = 'Renta_per_capita')
plt.title('Población VS renta disponible per cápita')
plt.tight_layout();
```

```python
# Como habiamos dicho antes con los numero, hay una relación psitiva
```
